1. SUMAMOS EL TOTAL DE IMPORTES DE CADA BENEFICIARIO

```python
sum_importe_benefeciario=compras_df.groupby('BENEFICIARIO').sum()['IMPORTE']
sum_importe_benefeciario
```

```python
sum_importe_benefeciario.reset_index().sort_values(by='IMPORTE', ascending=False).iloc [0:10].plot(kind='bar',x='BENEFICIARIO', y='IMPORTE')
```

```python
mean_import_sheet=compras_df.groupby('SHEET').mean()['IMPORTE']
mean_import_sheet.reset_index().sort_values(by='IMPORTE', ascending=False).iloc [0:10].plot(kind='bar', x='SHEET',y='IMPORTE')
```

```python
sum_import_sheet=compras_df.groupby('FECHA').sum()['IMPORTE']
sum_import_sheet.reset_index().sort_values(by='IMPORTE', ascending=False).iloc [0:10].plot(kind='bar', x='FECHA',y='IMPORTE')
```

## ANALISIS DE TEXTO

```python
import nltk 
nltk.download('book')
from nltk.book import *
from nltk.corpus import stopwords
print(stopwords.words('spanish'))
```

```python
textos_compras = ''
for row in compras_df['CONCEPTO']: 
    textos_compras = textos_compras + ' ' + row
```

```python
stopwords_list = stopwords.words('spanish') + ['2020', 'cont', 'millar', 'dpp', 'est', 'op', 'ret', 'mty', 'icom']
```

```python
textos_compras = ' '.join([x for x in textos_compras.strip().lower().split(' ') if x not in stopwords_list and len(x)>4])
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
wc = WordCloud(background_color= 'white')
wc.generate(textos_compras)

plt.axis("off")
plt.imshow(wc, interpolation='bilinear')
plt.show()
```
